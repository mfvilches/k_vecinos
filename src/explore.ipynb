```python
# Your code here

import pandas as pd

movies_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv')
credits_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv')
```

```python
import sqlite3

conn = sqlite3.connect("../data/raw/movies_db.db")

```

```python
movies_df.to_sql('tabla_movies',conn,if_exists='replace',index=False)
credits_df.to_sql('tabla_creditos',conn,if_exists='replace',index=False)
```

```python
print(movies_df.columns)
print(credits_df.columns)
```

```python
query = """
select tc.movie_id, tm.title, tm.overview, tm.genres, tm.keywords, tc.cast, tc.crew
from tabla_movies as tm
inner join tabla_creditos as tc
on tm.title = tc.title;
"""

df = pd.read_sql_query(query,conn)
conn.close()

df
```

```python
import json

df['genres'] = df['genres'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
```

```python
df
```

```python
df['keywords'] = df['keywords'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
df['cast'] = df['cast'].apply(lambda x:[item['name'] for item in json.loads(x)][:3] if pd.notna(x) else None)
```

```python
df
```

```python
df['crew'][0]
```

```python
df['crew'] = df['crew'].apply(lambda x: [item['name'] for item in json.loads(x) if item['job'] == 'Director'])
```

```python
df
```

```python
df['overview'] = df['overview'].apply(lambda x:[x])
```

```python
df
```

```python
df['overview'] = df['overview'].apply(lambda x: [str(x)])
df['genres'] = df['genres'].apply(lambda x: [str(element) for element in x])
df['keywords'] = df['keywords'].apply(lambda x: [str(element) for element in x])
df['cast'] = df['cast'].apply(lambda x: [str(element) for element in x])
df['crew'] = df['crew'].apply(lambda x: [str(element) for element in x])




df['tags'] = df['overview']+df['genres'] + df['keywords'] + df['cast'] + df['crew']
df['tags'] = df['tags'].apply(lambda x: ",".join(x).replace(","," "))
df
```

```python
df['tags'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['tags'])

model = NearestNeighbors(n_neighbors=6,algorithm='brute',metric='cosine')
model.fit(tfidf_matrix)

def obtener_recomendaciones(titulo_movie):
    indice_pelicula = df[df['title']==titulo_movie].index[0]
    distancias, indices = model.kneighbors(tfidf_matrix[indice_pelicula])
    peliculas_parecidas = [(df['title'][i],distancias[0][j]) for j,i in enumerate(indices[0])]
    return peliculas_parecidas[1:]

pelicula_consultar = "How to Train Your Dragon"
recomendaciones = obtener_recomendaciones(pelicula_consultar)
for pelicula, distancia in recomendaciones:
    print("-Pelicula {}".format(pelicula))

```
